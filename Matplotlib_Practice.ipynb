```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,7,9,14]) # if you provide a single list, it is assumed to be y values
```

```python
plt.plot([1,2,7,8],[1,8,7,14])
```

```python
# Default is 'b-' for solid blue line
plt.plot([1,2,7,8],[1,8,7,14], 'ro')
```

```python
# numpy arrays
a = np.array([[1,2,3,4],[5,6,7,8]])

plt.plot(a[0], a[1], 'go')
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(a[0], a[1], 'b--')
```

```python
# numpy function to create an array of 20 evenly spaced values between 0 and 5
t = np.linspace(0, 5, 20)

plt.plot(t, t**2, color='coral', linestyle='--')
plt.plot(t, t*4, color='indigo', linewidth=2)
```

```python
# subplot to add more than one plot in one figure
# parameters are number of rows, number of columns, index
plt.subplot(1, 3, 1)
plt.plot(t, t, 'c--')

plt.subplot(1, 3, 2)
plt.plot(t, t**2, 'b-')

plt.subplot(1, 3, 3)
plt.plot(t, t**3, 'mo')
```

```python
# Bar graphs
grades = {'John': 70,
          'Susan': 82,
          'Ben': 92,
          'Alex': 76,
          'David': 97}

plt.bar(range(len(grades)), list(grades.values()), color='darkturquoise')
plt.xticks(range(len(grades)), list(grades.keys()))
plt.xlabel('Students')
plt.ylabel('Grades')
```

```python
# Histograms

x = np.random.randn(1000) #generate random data

plt.hist(x, color='seagreen')
```
