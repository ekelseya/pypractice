```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Read csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='unicode-escape')

# Print first five records
df.head()
```

```python
# scatterplot

sns.lmplot(x='Attack', y='Defense', data=df)

plt.show()
```

```python
# scatterplot with extra arguments (no regression line and Stage shown by hue)

sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue="Stage")

plt.show()
```

```python
# Boxplot

df_copy = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

sns.boxplot(data=df_copy)
```

```python
# Violin plots

sns.violinplot(data=df_copy)

plt.show()
```

```python
# Violin comparing variance Type and Attack

plt.figure(figsize=(10,6))

sns.violinplot(x='Type 1', y='Attack', data=df)

plt.show()
```

```python
# Heatmap

corr = df_copy.corr() # create correlation

sns.heatmap(corr)
```

```python
# Histogram

sns.distplot(df.Attack, color='blue') # distplot is deprecated
```

```python
sns.displot(df.Attack, kde=True, color='blue') 
```

```python
# Count plots

sns.countplot(x='Type 1', data=df)
plt.xticks(rotation=-45)
```

```python
# Density plot - distribution between 2 variables

sns.jointplot(x=df.Attack, y=df.Defense, kind='kde')
```
